# Cannocial correlation anaylsis

## Theory

### Quick overview

CCA is a method for finding the correlation relationship between two or more multidimesnional datasets, similar to PCA/partial least squares regression. However, it finds a canonical co-ordinate space that maximizes correlations between projections of the datasets onto that space.

CCA can handle multiple different datasets of different sizes and doesn't presuppose any relationship between datasets like regression (i.e no independent/dependent variables). It also measures a relationship between datasets that is beyond similairty (which is what correlation analysis does).

## Maths

Given two zero-mean datasets (X, Y) CCA finds a canonical coordinate space that maximizes correlations between projects of datasets onto that space.

For each dimensions of the coordinate space are a pair of projections aj = (a1j, a2j, ‚Ä¶anj) and bj = (b1j, b2j, ‚Ä¶bmj)
These are called canonical weights


The projections of the datasets X and Y onto each dimension of the canonical space are a pair of d-dimensional vectors (inner dot product of canonical weights and dataset) 

uj = ‚å©aj, X‚å™and vj = ‚å©bj, Y‚å™

$$
pj = \max \frac{\langle u_j, v_j \rangle}{\|u_j\| \|v_j\|}
$$
CCA maximizes the correlations between each pair of canonical components

CCA is solved by iterating over the pair of canonical components Ux and Vx to maximise Px (i.e p1 = corr(v1, u1), p2= corr(v2, p2) etc) with the total number of pairs being contstrained to the minium number of dimensions in each dataset.

However in practice it is formulated as a generalized eigenvalue problem in one shot. To do this the function that solves for the maxmimum of the CCA vector is rewritten as a series of covariance matricies Cxx and Cyy (autocovariance matricies) and Cxy (sample covairnace matrices)

### Maths and steps in python

```python
# Fake data
np.random.seed(42)
X = np.random.rand(100, 5)  # 100 samples, 5 variables
Y = np.random.rand(100, 3)
```

1. Center the data

```python
X -= np.mean(X, axis=0)
Y -= np.mean(Y, axis=0)
```

2. Compute the covariance matrices

```python
n = X.shape[0]
Cxx = np.cov(X, rowvar=False)
Cyy = np.cov(Y, rowvar=False)
Cxy = np.cov(X, Y, rowvar=False)[:X.shape[1], X.shape[1]:]
```

3. Solve the eigenvalue

```python
inv_Cxx = np.linalg.pinv(Cxx)
inv_Cyy = np.linalg.pinv(Cyy)
eig_matrix = inv_Cxx @ Cxy @ inv_Cyy @ Cxy.T
eigvals, eigvecs = np.linalg.eigh(eig_matrix)
```

4. Order the eigvals

```python
idx = np.argsort(eigvals)[::-1]
eigvals = eigvals[idx]
eigvecs = eigvecs[:, idx]
```

4. Calculate the weight

```python
A = eigvecs
B = inv_Cyy @ Cxy.T @ A
```

5. Canonical correlations (square root of eigenvalues)

```python
num_correlations = min(X.shape[1], Y.shape[1])
corr = np.sqrt(np.abs(eigvals[:num_correlations]))
```

## CA3

$$
min_{\vec{w}_{x_1}, \vec{w}_{x_2}, \vec{w}_a, \vec{w}_b} \left( -\vec{w}_{x_1}^T S_{xa} \vec{w}_a - \vec{w}_{x_2}^T S_{xb} \vec{w}_b + \sum_{i \in \{x_1, x_1, a, b\}} \frac{1}{2} \lambda_i \left( \vec{w}_i^T S_{ii} \vec{w}_i - 1 \right) + \theta_{rr}(\vec{w}_{x_1}, \vec{w}_{x_2}) \right)
$$

```python
import numpy as np
from scipy.optimize import minimize

def objective_function(w, S_xa, S_xb, S_ii_list, lambda_list, theta_r):
    """
    Objective function to be minimized.

    Args:
        w: Concatenated weight vectors [wx1, wx2, wa, wb].
        S_xa: Covariance matrix between neuroimaging (dataset 1) and behavioral (dataset 1).
        S_xb: Covariance matrix between neuroimaging (dataset 2) and behavioral (dataset 2).
        S_ii_list: List of covariance matrices [S_x1x1, S_x2x2, S_aa, S_bb].
        lambda_list: List of Lagrange multipliers [lambda_x1, lambda_x2, lambda_a, lambda_b].
        theta_r: Hyperparameter controlling dissimilarity penalty.

    Returns:
        Value of the objective function.
    """

    # Extract weight vectors from the concatenated vector 'w'.
    dim_x1 = S_xa.shape[0]
    dim_x2 = S_xb.shape[0]
    dim_a = S_xa.shape[1]
    dim_b = S_xb.shape[1]

    wx1 = w[:dim_x1]
    wx2 = w[dim_x1:dim_x1 + dim_x2]
    wa = w[dim_x1 + dim_x2:dim_x1 + dim_x2 + dim_a]
    wb = w[dim_x1 + dim_x2 + dim_a:]

    # Calculate correlation terms.
    term1 = -np.dot(wx1.T, np.dot(S_xa, wa))
    term2 = -np.dot(wx2.T, np.dot(S_xb, wb))

    # Calculate regularization terms.
    term3 = 0
    for i, S_ii in enumerate(S_ii_list):
        term3 += 0.5 * lambda_list[i] * (np.dot(w[sum([S_ii_list[j].shape[0] for j in range(i)]):sum([S_ii_list[j].shape[0] for j in range(i+1)])].T, np.dot(S_ii, w[sum([S_ii_list[j].shape[0] for j in range(i)]):sum([S_ii_list[j].shape[0] for j in range(i+1)])])) - 1)

    # Calculate dissimilarity penalty (example using L2 norm).
    delta_w = wx1 - wx2
    term4 = theta_r * 0.5 * np.linalg.norm(delta_w)**2

    return term1 + term2 + term3 + term4

def solve_generalized_cca(S_xa, S_xb, S_ii_list, lambda_list, theta_r):
    """
    Solves the generalized CCA problem.

    Args:
        S_xa: Covariance matrix between neuroimaging (dataset 1) and behavioral (dataset 1).
        S_xb: Covariance matrix between neuroimaging (dataset 2) and behavioral (dataset 2).
        S_ii_list: List of covariance matrices [S_x1x1, S_x2x2, S_aa, S_bb].
        lambda_list: List of Lagrange multipliers [lambda_x1, lambda_x2, lambda_a, lambda_b].
        theta_r: Hyperparameter controlling dissimilarity penalty.

    Returns:
        Optimization result containing the weight vectors.
    """
    # Initial guess for the weight vectors.
    w_initial = np.random.rand(sum([S_ii_list[i].shape[0] for i in range(4)]))

    # Minimize the objective function.
    result = minimize(objective_function, w_initial, args=(S_xa, S_xb, S_ii_list, lambda_list, theta_r), method='L-BFGS-B')
    return result

# Example usage (replace with your actual covariance matrices and parameters):
dim_x1 = 100
dim_x2 = 100
dim_a = 50
dim_b = 50

S_xa = np.random.rand(dim_x1, dim_a)
S_xb = np.random.rand(dim_x2, dim_b)
S_x1x1 = np.random.rand(dim_x1, dim_x1)
S_x2x2 = np.random.rand(dim_x2, dim_x2)
S_aa = np.random.rand(dim_a, dim_a)
S_bb = np.random.rand(dim_b, dim_b)

S_ii_list = [S_x1x1, S_x2x2, S_aa, S_bb]
lambda_list = [1.0, 1.0, 1.0, 1.0]  # Example lambda values
theta_r = 1.0  # Example theta_r value

result = solve_generalized_cca(S_xa, S_xb, S_ii_list, lambda_list, theta_r)

if result.success:
    print("Optimization successful!")
    print("Weight vectors:", result.x)
else:
    print("Optimization failed:", result.message)
```

```python
import numpy as np
from scipy.optimize import minimize

# Define the C3A objective function
def c3a_objective(params, x1, x2, a, b, S_xa, S_xb, S_aa, S_bb, S_xx1, S_xx2, lambda_vals, theta_r, r_func):
    # Dimensions of the datasets
    n_x1 = x1.shape[1]  # Number of features in x1
    n_x2 = x2.shape[1]  # Number of features in x2
    n_a = a.shape[1]    # Number of behavioral features in a
    n_b = b.shape[1]    # Number of behavioral features in b
    
    # Unpack parameters: wx1, wx2, wa, wb
    wi_x1 = params[:n_x1]
    wi_x2 = params[n_x1:n_x1 + n_x2]
    wi_a = params[n_x1 + n_x2:n_x1 + n_x2 + n_a]
    wi_b = params[n_x1 + n_x2 + n_a:n_x1 + n_x2 + n_a + n_b]
    
    # First two terms (cross-correlation)
    term1 = np.dot(wi_x1.T, np.dot(S_xa, wi_a))
    term2 = np.dot(wi_x2.T, np.dot(S_xb, wi_b))
    
    # Regularization term
    reg_term1 = 0.5 * lambda_vals[0] * np.dot(wi_x1.T, np.dot(S_xx1, wi_x1))
    reg_term2 = 0.5 * lambda_vals[1] * np.dot(wi_x2.T, np.dot(S_xx2, wi_x2))
    reg_term3 = 0.5 * lambda_vals[2] * np.dot(wi_a.T, np.dot(S_aa, wi_a))
    reg_term4 = 0.5 * lambda_vals[3] * np.dot(wi_b.T, np.dot(S_bb, wi_b))
    
    # Cross-dataset weight similarity penalty
    dissimilarity_term = theta_r * r_func(wi_x1, wi_x2)
    
    # Total objective function
    objective_value = -(term1 + term2) + reg_term1 + reg_term2 + reg_term3 + reg_term4 + dissimilarity_term
    
    return objective_value

# Define the dissimilarity penalty function (penalizes the difference between wi_x1 and wi_x2)
def r(wi_x1, wi_x2):
    return 0.5 * np.linalg.norm(wi_x1 - wi_x2)**2

# Function to compute CCA components and correlation coefficient
def compute_cca_components(x1, x2, a, b, lambda_vals, theta_r):
    # Compute covariance matrices
    S_xa = np.cov(x1.T, a.T)  # Covariance between x1 and a
    S_xb = np.cov(x2.T, b.T)  # Covariance between x2 and b
    S_aa = np.cov(a.T, a.T)    # Covariance between a and a
    S_bb = np.cov(b.T, b.T)    # Covariance between b and b
    S_xx1 = np.cov(x1.T, x1.T)  # Covariance of x1 with itself
    S_xx2 = np.cov(x2.T, x2.T)  # Covariance of x2 with itself
    
    # Initialize random parameters for wi_x1, wi_x2, wi_a, wi_b
    n_x1 = x1.shape[1]
    n_x2 = x2.shape[1]
    n_a = a.shape[1]
    n_b = b.shape[1]
    
    # Random initial weights
    init_params = np.concatenate([np.random.randn(n_x1), np.random.randn(n_x2), 
                                  np.random.randn(n_a), np.random.randn(n_b)])
    
    # Minimize the C3A objective function
    result = minimize(c3a_objective, init_params, args=(x1, x2, a, b, S_xa, S_xb, S_aa, S_bb, S_xx1, S_xx2, lambda_vals, theta_r, r),
                      method='BFGS')

    # Extract the optimized weight vectors
    optimized_weights = result.x
    wi_x1 = optimized_weights[:n_x1]
    wi_x2 = optimized_weights[n_x1:n_x1 + n_x2]
    wi_a = optimized_weights[n_x1 + n_x2:n_x1 + n_x2 + n_a]
    wi_b = optimized_weights[n_x1 + n_x2 + n_a:n_x1 + n_x2 + n_a + n_b]
    
    # Compute the projections of the datasets
    x1_projected = np.dot(x1, wi_x1)
    x2_projected = np.dot(x2, wi_x2)
    
    # Compute the canonical correlation coefficient between x1 and x2
    corr_coeff = np.corrcoef(x1_projected, x2_projected)[0, 1]
    
    return wi_x1, wi_x2, wi_a, wi_b, corr_coeff

# Example data: Replace these with your actual data
np.random.seed(42)
n_samples = 100
n_features_x1 = 60
n_features_x2 = 60
n_features_a = 10
n_features_b = 10

# Generate random example data
x1 = np.random.randn(n_samples, n_features_x1)
x2 = np.random.randn(n_samples, n_features_x2)
a = np.random.randn(n_samples, n_features_a)
b = np.random.randn(n_samples, n_features_b)

# Regularization parameters and penalty strength
lambda_vals = np.array([1.0, 1.0, 1.0, 1.0])
theta_r = 1.0

# Compute CCA components and the correlation coefficient
wi_x1, wi_x2, wi_a, wi_b, corr_coeff = compute_cca_components(x1, x2, a, b, lambda_vals, theta_r)

# Output the results
print("Optimized CCA weights:")
print("wi_x1:", wi_x1)
print("wi_x2:", wi_x2)
print("wi_a:", wi_a)
print("wi_b:", wi_b)
print(f"Canonical correlation coefficient: {corr_coeff}")
```
